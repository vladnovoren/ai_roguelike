```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def CalcUtilityImg(x_max, y_max, utility):
    x_grid = np.arange(0, x_max, 0.01)
    y_grid = np.arange(0, y_max, 0.01)

    img = np.zeros((y_grid.size, x_grid.size))

    for y, y_val in enumerate(y_grid):
        for x, x_val in enumerate(x_grid):
            img[-x][y] = utility(x_val, y_val)

    return img
```

```python
def MoveToBaseUtility(ally_dist, base_dist):
    return  2 * (ally_dist ** 2 + base_dist ** 2)
    return ally_dist * base_dist
    return 4 * max(0, 5 - ally_dist) * max(0, 5 - base_dist)

def AttackEnemyUtility(enemy_dist, base_dist):
    return max(0, 100 - (enemy_dist * base_dist) ** 4)
    return max(0, 75 - enemy_dist ** 2 - base_dist ** 2)
```

```python
img = CalcUtilityImg(5, 5, MoveToBaseUtility)

fig, ax = plt.subplots(1, 2)
fig.suptitle("Move to base utility")
fig.set_size_inches(13, 5)
im = ax[0].imshow(img, extent=[0, 5, 0, 5], cmap='Greys')
ax[0].set_aspect('auto')
ax[0].set_xlabel("ally_dist")
ax[0].set_ylabel("base_dist")
plt.colorbar(im)

im = ax[1].imshow(img > 50, extent=[0, 5, 0, 5], cmap='Greys')
ax[1].set_aspect('auto')
ax[1].set_xlabel("ally_dist")
ax[1].set_ylabel("base_dist")
plt.colorbar(im)
```

```python
img = CalcUtilityImg(5, 5, AttackEnemyUtility)

fig, ax = plt.subplots(1, 2)
fig.suptitle("Attack enemy utility")
fig.set_size_inches(13, 5)
im = ax[0].imshow(img, extent=[0, 5, 0, 5], cmap='Greys')
ax[0].set_aspect('auto')
ax[0].set_xlabel("enemy_dist")
ax[0].set_ylabel("base_dist")
plt.colorbar(im)

im = ax[1].imshow(img > 50, extent=[0, 5, 0, 5], cmap='Greys')
ax[1].set_aspect('auto')
ax[1].set_xlabel("enemy_dist")
ax[1].set_ylabel("base_dist")
plt.colorbar(im)
```
